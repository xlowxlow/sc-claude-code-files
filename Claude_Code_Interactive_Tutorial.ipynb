## ğŸ” æ™ºèƒ½æœç´¢åŠŸèƒ½ | Smart Search Function

ä½¿ç”¨ä¸‹é¢çš„æœç´¢å·¥å…·å¿«é€Ÿæ‰¾åˆ°æ‚¨éœ€è¦çš„Claude CodeçŸ¥è¯†ï¼

Use the search tool below to quickly find the Claude Code knowledge you need!

```python
# ğŸ” æ™ºèƒ½æœç´¢åŠŸèƒ½ | Smart search function
def search_knowledge(query: str, category: str = "all", language: str = "both") -> List[Dict]:
    """
    æœç´¢Claude CodeçŸ¥è¯†åº“
    Search Claude Code knowledge base
    
    Args:
        query: æœç´¢å…³é”®è¯ | Search keyword
        category: ç±»åˆ«ç­›é€‰ | Category filter
        language: è¯­è¨€åå¥½ | Language preference
    """
    results = []
    query_lower = query.lower()
    
    # æœç´¢å‘½ä»¤ | Search commands
    if category in ["all", "commands"]:
        for cmd, info in claude_code_knowledge["commands"].items():
            if (query_lower in cmd.lower() or 
                query_lower in info["function_zh"].lower() or 
                query_lower in info["function_en"].lower() or
                query_lower in info["use_case_zh"].lower() or
                query_lower in info["use_case_en"].lower()):
                results.append({
                    "type": "å‘½ä»¤ | Command",
                    "name": cmd,
                    "info": info
                })
    
    # æœç´¢å¿«æ·é”® | Search shortcuts
    if category in ["all", "shortcuts"]:
        for shortcut, info in claude_code_knowledge["shortcuts"].items():
            if (query_lower in shortcut.lower() or 
                query_lower in info["function_zh"].lower() or 
                query_lower in info["function_en"].lower()):
                results.append({
                    "type": "å¿«æ·é”® | Shortcut",
                    "name": shortcut,
                    "info": info
                })
    
    # æœç´¢é«˜çº§åŠŸèƒ½ | Search advanced features
    if category in ["all", "advanced"]:
        for feature, info in claude_code_knowledge["advanced_features"].items():
            if (query_lower in feature.lower() or 
                query_lower in info["description_zh"].lower() or 
                query_lower in info["description_en"].lower()):
                results.append({
                    "type": "é«˜çº§åŠŸèƒ½ | Advanced Feature",
                    "name": feature,
                    "info": info
                })
    
    return results

def display_search_results(results: List[Dict], language: str = "both"):
    """
    æ˜¾ç¤ºæœç´¢ç»“æœ
    Display search results
    """
    if not results:
        print("âŒ æœªæ‰¾åˆ°ç›¸å…³ç»“æœ | No results found")
        return
    
    print(f"ğŸ¯ æ‰¾åˆ° {len(results)} ä¸ªç»“æœ | Found {len(results)} results\n")
    
    for i, result in enumerate(results, 1):
        print(f"\n{'='*50}")
        print(f"ğŸ“ ç»“æœ {i} | Result {i}: {result['type']}")
        print(f"ğŸ·ï¸ åç§° | Name: {result['name']}")
        
        info = result['info']
        
        if 'function_zh' in info:
            if language in ["both", "zh"]:
                print(f"ğŸ”§ åŠŸèƒ½: {info['function_zh']}")
            if language in ["both", "en"]:
                print(f"ğŸ”§ Function: {info['function_en']}")
                
        if 'use_case_zh' in info:
            if language in ["both", "zh"]:
                print(f"ğŸ’¡ ä½¿ç”¨åœºæ™¯: {info['use_case_zh']}")
            if language in ["both", "en"]:
                print(f"ğŸ’¡ Use Case: {info['use_case_en']}")
                
        if 'example' in info:
            print(f"ğŸ“ ç¤ºä¾‹ | Example: {info['example']}")
            
        if 'category' in info:
            print(f"ğŸ“‚ ç±»åˆ« | Category: {info['category']}")

print("âœ… æœç´¢åŠŸèƒ½å·²å‡†å¤‡å°±ç»ªï¼| Search function ready!")
```

### ğŸ® äº¤äº’å¼æœç´¢ç•Œé¢ | Interactive Search Interface

```python
# ğŸ® åˆ›å»ºäº¤äº’å¼æœç´¢ç•Œé¢ | Create interactive search interface

# æœç´¢è¾“å…¥æ¡† | Search input box
search_input = widgets.Text(
    placeholder="è¾“å…¥æœç´¢å…³é”®è¯... | Enter search keywords...",
    description="ğŸ” æœç´¢ | Search:",
    style={'description_width': 'initial'},
    layout=widgets.Layout(width='400px')
)

# ç±»åˆ«é€‰æ‹© | Category selection
category_dropdown = widgets.Dropdown(
    options=[
        ('å…¨éƒ¨ | All', 'all'),
        ('å‘½ä»¤ | Commands', 'commands'),
        ('å¿«æ·é”® | Shortcuts', 'shortcuts'),
        ('é«˜çº§åŠŸèƒ½ | Advanced', 'advanced')
    ],
    value='all',
    description="ğŸ“‚ ç±»åˆ« | Category:",
    style={'description_width': 'initial'}
)

# è¯­è¨€é€‰æ‹© | Language selection
language_dropdown = widgets.Dropdown(
    options=[
        ('ä¸­è‹±åŒè¯­ | Both', 'both'),
        ('ä¸­æ–‡ | Chinese', 'zh'),
        ('è‹±æ–‡ | English', 'en')
    ],
    value='both',
    description="ğŸŒ è¯­è¨€ | Language:",
    style={'description_width': 'initial'}
)

# æœç´¢æŒ‰é’® | Search button
search_button = widgets.Button(
    description="ğŸš€ æœç´¢ | Search",
    button_style='primary',
    tooltip="ç‚¹å‡»æœç´¢ | Click to search"
)

# è¾“å‡ºåŒºåŸŸ | Output area
output_area = widgets.Output()

def on_search_click(b):
    with output_area:
        output_area.clear_output()
        query = search_input.value.strip()
        if not query:
            print("âš ï¸ è¯·è¾“å…¥æœç´¢å…³é”®è¯ | Please enter search keywords")
            return
        
        print(f"ğŸ” æœç´¢: {query} | Searching: {query}")
        results = search_knowledge(query, category_dropdown.value, language_dropdown.value)
        display_search_results(results, language_dropdown.value)

# ç»‘å®šæœç´¢äº‹ä»¶ | Bind search event
search_button.on_click(on_search_click)

# æ”¯æŒå›è½¦æœç´¢ | Support Enter key search
def on_enter_pressed(change):
    if change['new'] and search_input.value.strip():
        on_search_click(None)

search_input.observe(on_enter_pressed, names='value')

# æ˜¾ç¤ºç•Œé¢ | Display interface
search_ui = widgets.VBox([
    widgets.HTML("<h3>ğŸ” Claude Code çŸ¥è¯†æœç´¢ | Knowledge Search</h3>"),
    widgets.HBox([search_input, search_button]),
    widgets.HBox([category_dropdown, language_dropdown]),
    output_area
])

display(search_ui)

# æ˜¾ç¤ºä½¿ç”¨æç¤º | Show usage tips
print("\nğŸ’¡ ä½¿ç”¨æç¤º | Usage Tips:")
print("- è¾“å…¥å…³é”®è¯å¦‚ 'init', 'è°ƒè¯•', 'debug' ç­‰ | Enter keywords like 'init', 'è°ƒè¯•', 'debug'")
print("- å¯ä»¥æœç´¢ä¸­æ–‡æˆ–è‹±æ–‡ | Can search in Chinese or English")
print("- æ”¯æŒæ¨¡ç³ŠåŒ¹é… | Supports fuzzy matching")
```
